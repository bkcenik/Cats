```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_data = pd.read_csv("weather.csv")
weather_data
Incidence_data=pd.read_csv("cleaned_incidence_rate_sorted.csv")
Incidence_data
merge_data=pd.merge(weather_data,Incidence_data,on='State',how='outer')
merge_data
```

```python
plt.scatter(weather_data["Inches"], Incidence_data["Incidence (Rate)"], marker="o")

plt.title("State Precipitation Inches vs. Incidence(Rate)")
plt.ylabel("Incidence(Rate)")
plt.xlabel("Precipitation Inches")
plt.grid(True)

#len(Incidence_data["Incidence (Rate)"]) == len(weather_data["Inches"])
```
